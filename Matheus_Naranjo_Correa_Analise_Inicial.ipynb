### Informações Gerais

Quantas linhas têm a tabela? Existe algum valor vazio que deveria ser tratado?

```python
def dataset_report(df: pd.DataFrame):
    print("--- Dataset Report ---")
    print(f"Número de Linhas: {df.shape[0]}")
    print(f"Número de Colunas: {df.shape[1]}\n")

    # Missing Values
    missing_total = df.isnull().sum().sum()
    print(f"Total de Células Vazias: {missing_total}\n")

    if missing_total > 0:
        # Columns with missing values
        missing_cols = df.isnull().sum()
        missing_cols = missing_cols[missing_cols > 0].sort_values(ascending=False)
        if not missing_cols.empty:
            print("Colunas com Valores Vazios:")
            for col, count in missing_cols.items():
                print(f"  - {col}: {count} valores vazios ({count/df.shape[0]:.2%})")
            print("")

        # Rows with missing values
        rows_with_missing = df[df.isnull().any(axis=1)]
        if not rows_with_missing.empty:
            print(f"Linhas com Valores Vazios: {rows_with_missing.shape[0]}")
    else:
        print("Não há valores vazios no dataset.\n")


    # Data Types
    print("Tipos de Dados por Coluna:")
    for col, dtype in df.dtypes.items():
        print(f"  - {col}: {dtype}")
    print("")

    # Memory Usage
    mem_usage = df.memory_usage(deep=True).sum()
    print(f"Tamanho do Dataset (Uso de Memória): {mem_usage / (1024**2):.2f} MB")
    print("--------------------")
```

```python
dataset_report(tabela)
```

```python
#Removendo linhas duplicadas e possíveis valores faltantes
tabela = tabela.dropna()
tabela = tabela[tabela['Economic Block'] !='Europe']
dataset_report(tabela)
```

### 1) Como foi a evolução das exportações para a França ao longo dos anos?

Dica: Crie uma tabela com os anos de 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
print("="*70)
print("EXPORTAÇÕES PARA A FRANÇA EM ANOS")
exportacoes_ano = tabela.groupby('Year')["US$ FOB"].sum()
print(exportacoes_ano.apply(lambda x: f"${x:,.0f}"))
print("="*70)
```

### 3) Em 2020 qual cidade mais exportou para a França?
Em valores US$

```python
tabela_2020 = tabela[tabela['Year'] == 2020]
tabela_2020.groupby('City')['US$ FOB'].sum().sort_values(ascending=False)
```

### 4) Quais os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em 2020) exportaram?

```python
produtos_mais_exportados = tabela_2020.copy()
maiores_cidades = tabela_2020.groupby('City')['US$ FOB'].sum().sort_values(ascending=False).head(2)
maiores_cidades = maiores_cidades.reset_index()


print("--- Produtos Mais Exportados pelas Maiores Cidades Exportadoras (2020) ---")

for cidade in maiores_cidades['City']:
    print(f"\n### Produtos mais exportados de {cidade}:")

    produtos_por_cidade = produtos_mais_exportados[produtos_mais_exportados['City'] == cidade]
    produtos_mais_exportados_cidade = produtos_por_cidade.groupby('SH4 Description')['US$ FOB'].sum().sort_values(ascending=False).head(2)

    # Converter o Series para um DataFrame para usar .style.format
    df_output = produtos_mais_exportados_cidade.reset_index()
    df_output.rename(columns={'US$ FOB': 'Valor US$ FOB'}, inplace=True)


    display(df_output.style.format({"Valor US$ FOB": "${:,.0f}"})
                        .set_properties(subset=['SH4 Description'], **{'text-align': 'left'}) # Alinha texto à esquerda
                        .set_table_styles([{'selector': 'th', 'props': [('text-align', 'left')]}] # Alinha cabeçalho à esquerda
                        ))
    print("-" * 60) # Linha separadora para melhorar a leitura entre as cidades
```
